```python
import sklearn 
from sklearn.datasets import load_iris 
from matplotlib import pyplot as plt 

#loading data 
iris=load_iris()

#the indices of the features that we are plotting 
a_index=0
b_index=1

# target names are in form of 1,2,3(1=setosa,2=versicolor,3=virgginica)
formatter=plt.FuncFormatter(lambda i,*args:iris.target_names[int(i)])

#size of graph 
plt.figure(figsize=(8,8))

#scatter(a,b,c=for diff. the 3 categories
plt.scatter(iris.data[:,a_index],iris.data[:,b_index],c=iris.target)

#it will show the colour bar indicating target names
plt.colorbar (ticks=[1,2,3],format=formatter)

plt.xlabel(iris.feature_names[a_index])
plt.ylabel(iris.feature_names[b_index])

plt.show()
```
