```python
import matplotlib.pyplot as plt
x=[2,4]
y=[4,5]
plt.xlabel("time")
plt.ylabel("speed")
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,5,9,0,4,5]
y=[2,5,8,0,1,7,3]

plt.xlabel("students")
plt.ylabel("tiffens")
plt.scatter(x,y,label="mines",marker='x',s=210,c='y')

plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,5,9,0,4,5]
y=[2,5,8,0,1,7,3]

x1=[3,4,5,6,7,8]
y1=[3,6,8,9,3,2]
plt.xlabel("students")
plt.ylabel("tiffens")
plt.scatter(x,y,marker='x',s=20,c='y')

plt.scatter(x1,y1,label="mines",s=20,c='b')
plt.show()
```
